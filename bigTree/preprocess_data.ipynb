```python
import pandas as pd
import json
from pandas import json_normalize
import datetime
import os
import requests

date = str(datetime.date.today())

curl =  "http://fetnet.infominer.io:8080/api/?start_tm=2022-01-01&end_tm=" + date
# curl =  "http://fetnet.infominer.io:8080"

# filename = i +"_"+date+".csv"

s=requests.get(curl, proxies={"http": "http://fetfw.fareastone.com.tw:8080"}).content
s = json.loads(s)
df = json_normalize(s) #Results contain the required data

df['date'] = pd.to_datetime(df['tm'], unit='s')

df_final = df[['id','date', 'url', 'title','content', 'source', 'location', 'type','tags']]

# data = pd.read_csv(io.StringIO(s.decode('utf-8')), sep=",")
save_path = "../data/"
filename = "data_" + date + ".csv"
path = os.path.join(save_path,filename) 
print(path)
df_final.to_csv(path, index = False)
# df = pd. read_json(curl, proxies={"http": "http://fetfw.fareastone.com.tw:8080"})
# print(df[['url','tm','title','content', 'source','location','type','tags','cht','aptg','']])

```
